```python
#Tasks 1: generate uniformly distributed random numbers
import numpy as np

# Set the seed to ensure reproducibility
np.random.seed()

# Generate 10 random numbers from uniform distribution [0, 5] with two decimals
random_numbers_u = np.random.uniform(0, 5, 10).round(2)

# Print the result
print("Generated 10 random numbers from U[0,5]:", random_numbers_u)
```

```python
#Task 2: generate normally distributed random numbers
import numpy as np
# Mean and vaciance for the normal distribution N(0,4)
mean_s = 0  # Mean
var_s = 4 # Variance

# Generate 10 random numbers with two decimal places from a normal distribution [0, 4]
random_numbers_s = np.random.normal(mean_s, var_s, 10)
random_numbers_s = np.round(random_numbers_s, 2)

# Print the generated random numbers
print("Generate 10 random numbers from N[0,4]:", random_numbers_s)
# Generate a bivariate normal distribution with mean 0 variance 4 for both X and Y and 0 covariance.
# Mean and standard deviation for the bivariate distribution
mean_b = [0, 0]  # Mean for both variables
cov_matrix_b = [[4, 0], [0, 4]]  # Covariance matrix for the two variables

# Generate 10 random numbers with two decimal places from a bivariate distribution [0, 4]
random_numbers_b = np.random.multivariate_normal(mean_b, cov_matrix_b, 10)
random_numbers_b = np.round(random_numbers_b, 2)

# Print the generated random numbers
print("Generate 10 random groups of numbers from N[[0 \\\ 0],[4,0 \\\ 0,4]]:")
for i in range(10):
  print(f'({random_numbers_b[i][0]}, {random_numbers_b[i][1]})', end=' ')
```

```python
#Task 4 Define functions determined by a data sample
import numpy as np
import matplotlib.pyplot as plt

# Sample data points (x, y)
x = np.array(random_numbers_u)
print("x:",x)
y = np.array(random_numbers_s)
print("y:",y)
# Define a function to fit the data (in this case, a polynomial)
def fit_polynomial(x, y, degree):
    coefficients = np.polyfit(x, y, degree)
    polynomial_function = np.poly1d(coefficients)
    return polynomial_function

# Fit a polynomial of degree 1 to the data
degree = 1
polynomial_function = fit_polynomial(x, y, degree)

# Generate x values for plotting
x_values = np.linspace(min(x) - 1, max(x) + 1, 400)
y_values = polynomial_function(x_values)

# Plot the data points
plt.scatter(x, y, color='blue', label='Data Points')

# Plot the fitted polynomial function
plt.plot(x_values, y_values, color='red', label=f'Fitted Polynomial (Degree {degree})')

# Set labels and title
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Fitted Polynomial to Data Points')
plt.legend()

# Show the plot
plt.grid(True)
plt.show()

print(f"Fitted Polynomial Function (Degree {1}):")
print(polynomial_function)
```

```python
#Task 5 Optimization of a given function
import numpy as np
from scipy import optimize
from scipy.optimize import minimize
import matplotlib.pyplot as plt
# Sample data points (x, y)
x = np.array(random_numbers_u)
print("x:",x)
y = np.array(random_numbers_s)
print("y:",y)
# Define a function to fit the data (in this case, a polynomial)
def fit_polynomial(x, y, degree):
    coefficients = np.polyfit(x, y, degree)
    objective_function = np.poly1d(coefficients)
    return objective_function

# Fit a polynomial of degree 2 to the data
degree = 2
objective_function = fit_polynomial(x, y, degree)

# Generate x values for plotting
x_values = np.linspace(min(x) - 1, max(x) + 1, 400)
y_values = objective_function(x_values)

# Plot the data points
plt.scatter(x, y, color='blue', label='Data Points')

# Plot the fitted polynomial function
plt.plot(x_values, y_values, color='red', label=f'Fitted Polynomial (Degree {degree})')

# Set labels and title
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Fitted Polynomial to Data Points')
plt.legend()

# Show the plot
plt.grid(True)
plt.show()

print(f"Fitted Polynomial Function (Degree {2}):")
print(objective_function)
# Optimization on a closed interval[a,b]
# Define the closed interval for optimization [a, b]
a = 0
b = 5
initial_guess = 2
# Find the minimum of the function within the closed interval

min_result = minimize(objective_function, x0=(a+b)/2, bounds=[(a, b)] )

# Find the maximum of the function within the closed interval
max_result = minimize(lambda x: -objective_function(x), x0=(a+b)/2, bounds=[(a, b)])

# Extract the optimized values of x and the function value for min and max
minimized_x = min_result.x[0]
minimized_value = min_result.fun

maximized_x = max_result.x[0]
maximized_value = -max_result.fun  # Revert the sign to get the actual maximum value

print("Minimum Value (x, f(x)):", (minimized_x, minimized_value))
print("Maximum Value (x, f(x)):", (maximized_x, maximized_value))
```
